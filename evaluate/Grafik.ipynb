```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

#verilerin yüklenmesi
data=pd.read_excel("D:\Tuvis\c25-d100_nok\grafik_valid.xlsx")

#veri tipinin floata çevrilmesi
data['mAP_05'] = data['mAP_05'].astype('float')
data['Precision'] = data['Precision'].astype('float')
data['Recall'] = data['Recall'].astype('float')
data['F1-Score'] = data['F1_Score'].astype('float')
data['IoU'] = data['IoU'].astype('float')
data['Loss'] = data['Loss'].astype('float')

#ilk 10 verinin gösterilmesi
data.head(100)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.Iteration,data.mAP_05)
plt.title("mAP_05_Test")
plt.xlabel("Iteration")
plt.legend(['mAP_05 = %84'])
plt.show()

##
plt.figure(figsize=(12,6))
plt.plot(data.Iteration,data.Precision, color = 'brown') #biz belirlemediğimiz sürece rengi otomatik kendisi verir.
plt.title("Precision_Test")
plt.xlabel("Iteration")
plt.legend(['Precision = %84'])
plt.show()

##
plt.figure(figsize=(12,6))
plt.plot(data.Iteration,data.Recall,color = 'green')
plt.title("Recall_Test")
plt.xlabel("Iteration")
plt.legend(['Recall = %84'])
plt.show()

##
plt.figure(figsize=(12,6))
plt.plot(data.Iteration,data.F1_Score, color = 'red')
plt.title("F1_Score_Test")
plt.xlabel("Iteration")
plt.legend(['F1_Score = %84'])
plt.show()

##
plt.figure(figsize=(12,6))
plt.plot(data.Iteration,data.IoU, color = 'purple')
plt.title("IoU_Test")
plt.xlabel("Iteration")
plt.legend(['IoU = %84'])
plt.show()
```
